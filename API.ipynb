```python
import requests
import numpy as np





url= 'https://api.aviationapi.com/v1/airports'
airports= ['DCA', 'IAD', 'BWI']
airport_managers= {}




for airport in airports:
    params = {'apt': airport}
    resp = requests.get(url, params=params)
    
    if resp.status_code == 200:
        airport_data = resp.json()
        manager= airport_data.get("manager")
        airport_managers[airport]= manager
        print (f"Manager for {airport}: {manager}")



get_ipython().run_line_magic('matplotlib', 'inline')
```

```python

import json
import random

import numpy as np
import pandas as pd
import networkx as nx

g = nx.Graph()

with open("imdb_movies_2000to2022.prolific.json", "r") as in_file:
    for line in in_file:
        
        this_movie = json.loads(line)
            
        
        for actor_id,actor_name in this_movie['actors']:
            g.add_node(actor_name, actor_id=actor_id)
            
        i = 0 
        for left_actor_id,left_actor_name in this_movie['actors']:
            for right_actor_id,right_actor_name in this_movie['actors'][i+1:]:

                if g.has_edge(left_actor_name, right_actor_name):
                    g[left_actor_name][right_actor_name]['weight']=+1
                
                else:
                    g.add_edge(left_actor_name, right_actor_name, weight=1)
                
                print(left_actor_name, "<->", right_actor_name)
                
            i += 1 
            
      

print("Nodes:", len(g.nodes))

actor_neighbors= sorted(g.degree, key=lambda x: x[1], reverse=True)
print("Top 20 actors with the most neighbors: ")
for actor, x in actor_neighbors[:20]:
    print(f"{actor}:{x} neighbors")
```
